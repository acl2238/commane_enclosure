```python
from IPython.display import display
from ipyfilechooser import FileChooser

fc = FileChooser('/media/acl2238/3332-6631/')
fc.filter_pattern = '*.csv'
display(fc)

print(fc.selected_filename)
```

```python
# Carbon Monoxide Plots

import csv
from matplotlib import pyplot as plt
import numpy as np

yvals = []
xvals = []
ylabel = None
xlabel = "Time Elapsed (seconds)"

with open('/media/acl2238/3332-6631/gas_2024-05-20_1454.csv', mode='r')as file:
    csvFile = csv.reader(file)
    for line in csvFile:
        if ylabel == None:
            ylabel = line[4]
            continue
        yvals.append(line[4]) # Monoxide
        #xvals.append(line[1]) # Time, although Matplotlib automatically assigns x values if only y's are passed in

print(yvals)
#print(xvals)

plt.plot(yvals)
plt.ylabel(ylabel)
plt.xlabel(xlabel)

# I suppose the next thing to do now is to add file selection functionality so that you don't have to type out the path lol

```
